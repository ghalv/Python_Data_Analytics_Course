## ðŸŸ¥ Determine Job Title (1.7.3)

```python
# Code Needed
job_experience = 4
sql_experience = 2
if job_experience < 2:
  skill_level = 'Junior'
elif job_experience < 6 and sql_experience == 1:
  skill_level = 'Intermediate'
elif job_experience < 6 and sql_experience >= 2:
  skill_level = 'Senior'
elif job_experience > 5 and sql_experience >= 3:
  skill_level = 'Lead'
else:
  skill_level = 'Data'

title = skill_level + ' Analyst'
print(title)
```
